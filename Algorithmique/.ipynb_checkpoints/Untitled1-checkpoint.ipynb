```python
%matplotlib inline
import matplotlib.pyplot as plt
from networkx import nx

def diviseur(n):
    G = nx.DiGraph()
    
    for i in range (2,n+1):
        G.add_node(i)
        
    for i in range (2,n+1):
        for j in range (2,i+1):
            if (i%j == 0):
                G.add_edge(i,j)

    nx.draw_networkx(G)
    plt.show()
    
diviseur(50)
```
